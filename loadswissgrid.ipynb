```python
import pandas as pd
from datetime import timedelta

# Chargement du DataFrame sg_2023 à partir du fichier Excel
sg_2023 = pd.read_excel("C:/Users/souha/Downloads/EnergieUebersichtCH-2023.xlsx",
                        sheet_name='Zeitreihen0h15', skiprows=[1], index_col=0).astype(int)

# Colonnes à conserver
cols = [
    'Summe endverbrauchte Energie Regelblock Schweiz\nTotal energy consumed by end users in the Swiss controlblock',
    'Summe produzierte Energie Regelblock Schweiz\nTotal energy production Swiss controlblock',
    'Summe verbrauchte Energie Regelblock Schweiz\nTotal energy consumption Swiss controlblock',
    'Verbundaustausch CH->AT\nCross Border Exchange CH->AT',
    'Verbundaustausch AT->CH\nCross Border Exchange AT->CH',
    'Verbundaustausch CH->DE\nCross Border Exchange CH->DE',
    'Verbundaustausch DE->CH\nCross Border Exchange DE->CH',
    'Verbundaustausch CH->FR\nCross Border Exchange CH->FR',
    'Verbundaustausch FR->CH\nCross Border Exchange FR->CH',
    'Verbundaustausch CH->IT\nCross Border Exchange CH->IT',
    'Verbundaustausch IT->CH\nCross Border Exchange IT->CH'
]

sg_2023 = sg_2023[cols]

# Suppression des index dupliqués
sg_2023 = sg_2023[~sg_2023.index.duplicated(keep='first')]

# Renommage des colonnes
mapping = {
    'Summe endverbrauchte Energie Regelblock Schweiz\nTotal energy consumed by end users in the Swiss controlblock': 'Consommation_CH',
    'Summe produzierte Energie Regelblock Schweiz\nTotal energy production Swiss controlblock': 'Production_CH',
    'Summe verbrauchte Energie Regelblock Schweiz\nTotal energy consumption Swiss controlblock': 'Consommation_Brut_CH',
    'Verbundaustausch CH->AT\nCross Border Exchange CH->AT': 'CH->AT',
    'Verbundaustausch AT->CH\nCross Border Exchange AT->CH': 'AT->CH',
    'Verbundaustausch CH->DE\nCross Border Exchange CH->DE': 'CH->DE',
    'Verbundaustausch DE->CH\nCross Border Exchange DE->CH': 'DE->CH',
    'Verbundaustausch CH->FR\nCross Border Exchange CH->FR': 'CH->FR',
    'Verbundaustausch FR->CH\nCross Border Exchange FR->CH': 'FR->CH',
    'Verbundaustausch CH->IT\nCross Border Exchange CH->IT': 'CH->IT',
    'Verbundaustausch IT->CH\nCross Border Exchange IT->CH': 'IT->CH'
}

sg_2023.rename(columns=mapping, inplace=True)
sg_2023.index = pd.to_datetime(sg_2023.index, dayfirst=True)

# Chargement du DataFrame sg à partir du fichier CSV
sg = pd.read_csv("D:/Ecodynelec/ecodynelec/data/SwissGrid_total.csv", index_col=0)
sg.index = pd.to_datetime(sg.index)

# Suppression des index dupliqués
if not sg.index.is_unique:
    sg = sg.reset_index().drop_duplicates(subset='index').set_index('index')

if not sg_2023.index.is_unique:
    sg_2023 = sg_2023.reset_index().drop_duplicates(subset='index').set_index('index')

# Concaténation des DataFrames
SwissGrid_Total = pd.concat([sg.loc[:(sg_2023.index[0] - timedelta(minutes=15))], sg_2023], axis=0)


```
